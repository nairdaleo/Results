### Define a standard size for the images to be recognized

```python
std_size = 96
```

### Define the pubfig_data class and the droneface_data class

```python
class pubfig_data:
    images = list()
    target = list()
    target_name = list()

    def images_by_name(self, name):
        return [self.images[i] for i in range(self.target.size) if self.target_name[self.target[i]] == name]
    
class droneface_data:
    images = list()
    target = list()
    target_name = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k']
    target_info = {
        'subjectID': list(),
        'cameraType': list(),
        'heightID': list(),
        'imageType': list(),
        'distanceID': list()
    }

    def images_by_subjectID(self, subjectID):
        return [self.images[i] for i in range(len(self.target)) if self.target_info['subjectID'][i] == subjectID]
    
    def images_by_heightID(self, heightID):
        return [self.images[i] for i in range(len(self.target)) if self.target_info['heightID'][i] == heightID]
    
    def images_by_subjectID_and_heightID(self, subjectID, heightID):
        return [self.images[i] for i in range(len(self.target)) if (self.target_info['heightID'][i] == heightID and self.target_info['subjectID'][i] == subjectID)]
    
def preprocessing(img):
    data = img_to_array(img)
    data = preprocess_input(data)
    
    return data

def generate_data_dict(altitude_keys, identity_keys):
    ret_dict = {}
    for altitude in altitude_keys:
        ret_dict[altitude] = {}
        for identity in identity_keys:
            ret_dict[altitude][identity] = list()
    
    return ret_dict
```

## Scan pubfig12 method

```python
def scan_pubfig(path):
    from os.path import isfile, isdir
    from os import listdir, sep
    from cv2 import imread, cvtColor, COLOR_BGR2RGB, resize
    from numpy import array, shape, reshape
    
    images = list()
    target = list()
    target_name = list()

    # Initialize the data to be read
    data = pubfig_data()

    # Add a path separator if not already there
    if not path.endswith(sep):
        path = path + sep

    if isdir(path):
        for label in listdir(path):
            # Check if it's a directory
            if isdir(path + label):
                # Add the label if it's a directory
                target_name.append(label)

                # Iterate to read all files in the directory

                for file in listdir(path + label):
                    # Skip hidden configuration files
                    if file.startswith("."):
                        continue;

                    # Grab the image from that path and make it grayscale. Resize to a standard size
                    img = resize(cvtColor(imread(path + label + sep + file), COLOR_BGR2RGB),(std_size,std_size))

                    # Append the image inside
                    images.append(img)
                    # Assign the current label to the target list
                    target.append(target_name.index(label))

    # convert the information to expected numpy arrays
    data.images = images
    data.target = target
    data.target_name = target_name

    return data
```

## Scan DroneFace method

```python
def scan_droneFace_faces(path):
    from os.path import isfile, isdir
    from os import listdir, sep
    from cv2 import imread, cvtColor, COLOR_BGR2RGB, resize
    from numpy import array, shape, reshape

    images = list()
    target = list()
    target_info = {
        'subjectID' : list(),
        'cameraType' : list(),
        'heightID' : list(),
        'imageType' : list(),
        'distanceID' : list()
    }
    
    # Initialize the data to be returned
    data = droneface_data()

    # Add a path separator if not already there
    if not path.endswith(sep):
        path = path + sep

    if isdir(path):
        for file in listdir(path):
            # Skip hidden configuration files
            if file.startswith("."):
                continue;

            # Grab the image from that path and make it grayscale
            img = resize(cvtColor(imread(path + sep + file), COLOR_BGR2RGB),(std_size,std_size))

            images.append(img)

            # grab the encoded info in the file name
            name_data = file.split('_')

            target.append(data.target_name.index(name_data[0]))
            target_info['subjectID'].append(name_data[0])
            target_info['cameraType'].append(name_data[1])
            target_info['heightID'].append(name_data[2])
            target_info['imageType'].append(name_data[3])
            target_info['distanceID'].append(name_data[4].split('.')[0])

    # convert the rest of the information to expected numpy arrays
    data.images = images
    data.target = target
    data.target_info = target_info
    
    return data
```

## Recognition through single classifiers

### ROC

```python
# initialize rate and score dictionaries for each dataset
fpr_pf = dict()
tpr_pf = dict()
auc_score_pf = dict()

fpr_df = dict()
tpr_df = dict()
auc_score_df = dict()

roc_plot_size = (8,8)

# set a color set
from itertools import cycle
colours = cycle(['blue', 'blueviolet', 'brown', 'burlywood', 'cadetblue',
                 'crimson', 'darkblue', 'darkcyan', 'darkmagenta', 
                 'darkorange', 'darksalmon', 'darkslateblue',
                 'darkslategray', 'deeppink', 'indigo', 'maroon'])

# plot variable method
def multiclass_roc(y_test, y_score):
    from sklearn.metrics import roc_curve, auc
    from sklearn.preprocessing import label_binarize
    
    from scipy import interp
    
    import numpy as np
    
    y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
    
    n_classes = y_test_bin.shape[1]
    
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    
    # Compute individual class ROC curves
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # Compute micro average
    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    # Compute macro average
    
    ## Aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    
    ## Interpolate all ROC curves
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])
            
    ## Average and compute AUC
    mean_tpr /= n_classes
    
    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    
    return fpr, tpr, roc_auc
    
```

### Shuffle and split training and test sets

```python
from sklearn.model_selection import train_test_split

import numpy as np
import cv2

# Change to grayscale and use whole image as feature
grays = [cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) for image in data_pubfig.images]
X_pf = np.array(grays).reshape(len(data_pubfig.target),std_size*std_size)
y_pf = np.array(data_pubfig.target)

# Train-test split
X_pf_train, X_pf_test, y_pf_train, y_pf_test = train_test_split(X_pf, y_pf, test_size=0.5, random_state=0)

# Do the same with droneface data
grays = [cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) for image in data_droneface.images]
X_df["all"] = np.array(grays).reshape(len(data_droneface.target), std_size*std_size)
y_df["all"] = np.array(data_droneface.target)

X_df_train, X_df_test, y_df_train, y_df_test = train_test_split(X_df["all"], y_df["all"], test_size=0.5, random_state=0)
```

### Calculate PCA

```python
# Number of PCA components to keep:
n_components = 150
from sklearn.decomposition import PCA

pca_pf = PCA(n_components=n_components, svd_solver='randomized', whiten=True).fit(X_pf_train)
pca_df = PCA(n_components=n_components, svd_solver='randomized', whiten=True).fit(X_df_train)

X_pf_train_pca = pca_pf.transform(X_pf_train)
X_pf_test_pca = pca_pf.transform(X_pf_test)

X_df_train_pca = pca_df.transform(X_df_train)
X_df_test_pca = pca_df.transform(X_df_test)
```

### SVM models
* SVC with linear kernel
* SVC with poly kernel
* SVC with rbf kernel
* SVC with sigmoid kernel

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {'C' : [1e3, 2e3, 5e3],
              'gamma' : [0.001, 0.005, 0.01, 0.1]}

scoring_metrics = {'Accuracy': 'accuracy', 
                   'Precision': 'precision_macro', 
                   'F1': 'f1_macro'}

clf_pf_svc = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid, scoring=scoring_metrics, refit='Accuracy', cv=5)
clf_pf_svc.fit(X_pf_train_pca, y_pf_train)
y_pf_svc_score = clf_pf_svc.decision_function(X_pf_test_pca)
print('PubFig12 | Best estimator found: {0}'.format(clf_pf_svc.best_estimator_))

clf_df_svc = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid, scoring=scoring_metrics, refit='Accuracy', cv=5)
clf_df_svc.fit(X_df_train_pca, y_df_train)
y_df_svc_score = clf_df_svc.decision_function(X_df_test_pca)
print('DroneFace | Best estimator found: {0}'.format(clf_df_svc.best_estimator_))
```

#### Evaluate model quality

```python
# Predict from the test set
y_pf_svc_pred = clf_pf_svc.predict(X_pf_test_pca)
y_df_svc_pred = clf_df_svc.predict(X_df_test_pca)
```

### Prediction gallery

```python
# Qualitative evaluation of the predictions using matplotlib
# from https://scikit-learn.org/stable/auto_examples/applications/plot_face_recognition.html
def plot_gallery(images, titles, h, w, n_row=3, n_col=4):
    """Helper function to plot a gallery of portraits"""
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.85)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
#     plt.tight_layout()

# plot the result of the prediction on a portion of the test set
# from https://scikit-learn.org/stable/auto_examples/applications/plot_face_recognition.html
def title(y_pred, y_test, target_names, i):
    pred_name = target_names[y_pred[i]].rsplit(' ', 1)[-1]
    true_name = target_names[y_test[i]].rsplit(' ', 1)[-1]
    return 'Pred: %s\nTrue:      %s' % (pred_name, true_name)
```

### Train the model

```python
from keras import backend as K
from keras.models import Model
from keras.layers import Input, Layer

# Input for anchor, positive and negative images
in_a = Input(shape=(std_size, std_size, 3))
in_p = Input(shape=(std_size, std_size, 3))
in_n = Input(shape=(std_size, std_size, 3))

# Output for anchor, positive and negative embedding vectors
# The nn4_small model instance is shared (Siamese network)
emb_a = nn4_small2(in_a)
emb_p = nn4_small2(in_p)
emb_n = nn4_small2(in_n)

class TripletLossLayer(Layer):
    def __init__(self, alpha, **kwargs):
        self.alpha = alpha
        super(TripletLossLayer, self).__init__(**kwargs)
    
    def triplet_loss(self, inputs):
        a, p, n = inputs
        p_dist = K.sum(K.square(a-p), axis=-1)
        n_dist = K.sum(K.square(a-n), axis=-1)
        return K.sum(K.maximum(p_dist - n_dist + self.alpha, 0), axis=0)
    
    def call(self, inputs):
        loss = self.triplet_loss(inputs)
        self.add_loss(loss)
        return loss

# Layer that computes the triplet loss from anchor, positive and negative embedding vectors
triplet_loss_layer = TripletLossLayer(alpha=0.2, name='triplet_loss_layer')([emb_a, emb_p, emb_n])

# Model that can be trained with anchor, positive negative images
nn4_small2_train = Model([in_a, in_p, in_n], triplet_loss_layer)
```

## Prep the datasets

```python
import numpy as np
import os.path

class IdentityMetadata():
    def __init__(self, base, name, file):
        # dataset base directory
        self.base = base
        # identity name
        self.name = name
        # image file name
        self.file = file

    def __repr__(self):
        return self.image_path()

    def image_path(self):
        return os.path.join(self.base, self.name, self.file) 
    
def load_metadata(path):
    metadata = []
    for i in os.listdir(path):
        if i.startswith('.'):
            continue;
        for f in os.listdir(os.path.join(path, i)):
            # Check file extension. Allow only jpg/jpeg' files.
            if f.startswith('.'):
                continue;
            ext = os.path.splitext(f)[1]
            if ext == '.jpg' or ext == '.jpeg':
                metadata.append(IdentityMetadata(path, i, f))
    return np.array(metadata)
```

### Find the best threshold for verification

```python
from sklearn.metrics import f1_score, accuracy_score

distances = [] # squared L2 distance between pairs
identical = [] # 1 if same identity, 0 otherwise

num = len(metadata_pf)

for i in range(num - 1):
    for j in range(1, num):
        distances.append(distance(embedded_pf[i], embedded_pf[j]))
        identical.append(1 if metadata_pf[i].name == metadata_pf[j].name else 0)
        
distances = np.array(distances)
identical = np.array(identical)

thresholds = np.arange(0.9*min(distances), 1.1*max(distances), 0.01)

f1_scores = [f1_score(identical, distances < t) for t in thresholds]
acc_scores = [accuracy_score(identical, distances < t) for t in thresholds]

opt_idx = np.argmax(f1_scores)
# Threshold at maximal F1 score
opt_tau = thresholds[opt_idx]
# Accuracy at maximal F1 score
opt_acc = accuracy_score(identical, distances < opt_tau)

# Plot F1 score and accuracy as function of distance threshold
plt.plot(thresholds, f1_scores, label='F1 score');
plt.plot(thresholds, acc_scores, label='Accuracy');
plt.axvline(x=opt_tau, linestyle='--', lw=1, c='lightgrey', label='Threshold')
plt.title('PubFig12 | Accuracy at threshold {opt_tau:.2f} = {opt_acc:.3f}'.format(opt_tau=opt_tau, opt_acc=opt_acc));
plt.xlabel('Distance threshold')
plt.legend();
```

```python
from sklearn.metrics import f1_score, accuracy_score

distances = [] # squared L2 distance between pairs
identical = [] # 1 if same identity, 0 otherwise

num = len(metadata_df)

for i in range(num - 1):
    for j in range(1, num):
        distances.append(distance(embedded_df[i], embedded_df[j]))
        identical.append(1 if metadata_df[i].name == metadata_df[j].name else 0)
        
distances = np.array(distances)
identical = np.array(identical)

thresholds = np.arange(0.9*min(distances), 1.1*max(distances), 0.01)

f1_scores = [f1_score(identical, distances < t) for t in thresholds]
acc_scores = [accuracy_score(identical, distances < t) for t in thresholds]

opt_idx = np.argmax(f1_scores)
# Threshold at maximal F1 score
opt_tau = thresholds[opt_idx]
# Accuracy at maximal F1 score
opt_acc = accuracy_score(identical, distances < opt_tau)

# Plot F1 score and accuracy as function of distance threshold
plt.plot(thresholds, f1_scores, label='F1 score');
plt.plot(thresholds, acc_scores, label='Accuracy');
plt.axvline(x=opt_tau, linestyle='--', lw=1, c='lightgrey', label='Threshold')
plt.title('DroneFace | Accuracy at threshold {opt_tau:.2f} = {opt_acc:.3f}'.format(opt_tau=opt_tau, opt_acc=opt_acc));
plt.xlabel('Distance threshold')
plt.legend();
```

# Face detection
* Using cascaded haar detector
* Using dlib's HOG
* Using MTCNN

```python
import dlib

# OpenCV cascaded haar detector
face_cascade = cv2.CascadeClassifier('cascades/haarcascade_frontalface_default.xml')

# Dlib's HOG feature-extractor
dlib_detector = dlib.get_frontal_face_detector()
dlib_predictor = dlib.shape_predictor('models/shape_predictor_68_face_landmarks.dat')

# MTCNN
from mtcnn.mtcnn import MTCNN
mtcnn_detector = MTCNN()
```

```python
# dlib landmark ranges
landmark_ranges = list()

# Chin: 1-17
landmark_ranges.append(range(1,17))

# Eyebrow 1: 18 to 22
landmark_ranges.append(range(18,22))

# Eyebrow 2: 23 to 27
landmark_ranges.append(range(23,27))

# Nose bridge; 28 to 31
landmark_ranges.append(range(28,31))

# Nose arc: 32 to 36
landmark_ranges.append(range(32,36))

# Eye 1: 37 to 42
landmark_ranges.append(range(37,42))

# Eye 2: 43 to 48
landmark_ranges.append(range(43,48))

# Upper lip 1: 49 to 55
landmark_ranges.append(range(49,55))

# Lower lip 1: 56 to 60
landmark_ranges.append(range(56,60))

# Upper lip 2: 61 to 65
landmark_ranges.append(range(61,65))

# Lower lip 2: 66 to 68
landmark_ranges.append(range(66,68))

def extract_faces(img, show_detection=False, img_is_color=True, detector='opencv'):
    if img_is_color:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else:
        gray = img
    
    if detector == 'opencv':

        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)

        if show_detection:
            plt.figure(figsize=(4,4))
            if len(faces) > 0:
                
                plt.imshow(img)
                
                for (x, y, w, h) in faces:
                    plt.gca().add_patch(patches.Rectangle((x,y), w, h, fill=False, color='green'))
                    
                plt.title('Found {0} {1}'.format(len(faces), ('face' if len(faces) == 1 else 'faces')))    
                
            else:
                plt.imshow(img)
                plt.title('No faces found')

        return faces
    
    if detector == 'dlib':
        
        faces = dlib_detector(gray)
        
        landmarks_all = list()
        
        if show_detection:
            plt.figure(figsize=(4,4))
            plt.imshow(img)
            
        for face in faces:
            landmarks = dlib_predictor(gray, face)
            landmarks_all.append(landmarks)
                
            if show_detection:
                plt.gca().add_patch(patches.Rectangle((face.left(), face.top()), face.width(), face.height(), fill=False, color='pink'))
                
                for lr in landmark_ranges:
                    x = [landmarks.part(i-1).x for i in lr]
                    y = [landmarks.part(i-1).y for i in lr]
                    plt.plot(x, y, '-', markersize=2, color='pink')
            
                plt.title('Found {0} {1}'.format(len(faces), ('face' if len(faces) == 1 else 'faces'))) 

        return faces, landmarks_all
                                    
    if detector == 'mtcnn':
        faces = mtcnn_detector.detect_faces(img)
        
        if show_detection:
            plt.figure(figsize=(4,4))
            plt.imshow(img)
            
            for face in faces:
                x, y, w, h = face['box']
                plt.gca().add_patch(patches.Rectangle((x,y), w, h, fill=False, color='lightblue', linewidth=4))
                
                for keypoint in list(face['keypoints']):
                    plt.gca().add_patch(patches.Circle(face['keypoints'][keypoint], 1, color='lightblue'))
            
        return faces
```

```python
def live_predict_pf(path):
    plt.figure()
    
    example_image = load_image(path)   
    img = align_image(example_image)
    if img is None:
        img = cv2.resize(example_image[y:y+h, x:x+w],(std_size,std_size))
    bb = alignment.getLargestFaceBoundingBox(example_image)
    img = (img / 255.).astype(np.float32)
    
    example_emb = model_pf.predict(np.expand_dims(img, axis=0))[0]
    
    example_prediction = knn_pf.predict([example_emb])
    
    example_identity = encoder_pf.inverse_transform(example_prediction)[0]

    plt.imshow(example_image)
    plt.gca().add_patch(patches.Rectangle((bb.left(), bb.top()), bb.width(), bb.height(), fill=False, color='red'))
    plt.title('Recognized as {example_identity}'.format(example_identity=example_identity));
```

```python
def live_predict_multiple_pf(path, fs=None, method='facenet'):
    if fs is None:
        plt.figure()
    else:
        plt.figure(figsize=fs)
    
    example_image = load_image(path)
    
    faces = extract_faces(example_image, detector='mtcnn')
    plt.imshow(example_image)
    
    for face, colour in zip(faces, colours):
        x, y, w, h = face['box']
        plt.gca().add_patch(patches.Rectangle((x,y), w, h, fill=False, color=colour))
                
        for keypoint in list(face['keypoints']):
            plt.gca().add_patch(patches.Circle(face['keypoints'][keypoint], 0.8,color=colour))
            
        img = align_image(example_image[y:y+h, x:x+w])
        if img is None:
            img = cv2.resize(example_image[y:y+h, x:x+w],(std_size,std_size))
        
        if method == 'facenet':
            img = (img / 255.).astype(np.float32)
        
            example_emb = model_pf.predict(np.expand_dims(img, axis=0))[0]
            example_prediction = knn_pf.predict([example_emb])
            
            example_identity = encoder_pf.inverse_transform(example_prediction)[0]
        
        if method == 'mlp':
            # Change to grayscale and use whole image as feature
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            X = np.array(gray).reshape(1,std_size*std_size)

            example_prediction = clf_pf_mlp.predict(pca_pf.transform(X))[0]
        
            example_identity = data_pubfig.target_name[example_prediction]
        
        plt.text(int(0.9*x),int(0.9*y),'{example_identity}'.format(example_identity=example_identity), backgroundcolor='gray');
                        
    plt.title('Found {0} {1}'.format(len(faces), ('face' if len(faces) == 1 else 'faces')))
```
